#Creating the Linear Model#

Using statsmodels first, and then maybe scikit learn

```python
lm = smf.ols(formula='LV_AGE_DIFF0 ~ DISC_SCORE0', data=data).fit()
```

```python
lm.params
```

```python
lm.conf_int()
```

```python
lm.pvalues
```

```python
lm.rsquared
```

```python
lm2 = smf.ols(formula='LV_AGE_DIFF0 ~ DISC_SCORE0 + BMI0', data=data).fit()
print lm2.pvalues
```
